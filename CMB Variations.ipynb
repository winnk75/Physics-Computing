# Plotting Variations in the Cosmic Microwave Background


This notebook is a proof of concept that illustrates a portion of work I'm doing for my thesis. I use a modified version of [CAMB](http://camb.info) to generate CMB power spectra, and then compare them. The comparison is done using Markov chain Monte Carlo (MCMC), which is more or less a fancy word to say that we generate a bunch of different models and see which ones fit the data the best. Normally this process is handled by an algorithmic sampler, but I've broken down one small step here as a learning tool. In this notebook, we get under the hood to see how a sampling algorithm makes decisions, and we see one method by which Python can load and plot data.

For this notebook I've done a series of runs with Camb (inside [Cosmosis](https://bitbucket.org/joezuntz/cosmosis/overview)), varying just the Hubble parameter from the Planck best fit value of about $67 \ [  km \ s^{-1} \ Mpc^{-1}  ]$.

My own research uses a modified version of CAMB to probe non-standard cosmological parameters, but the theory remains exactly the same.

## Roadmap

First we break the project down into steps:

### Loading Observational Data from a FITS File

In the plot above, the curves are all over the place. Let's introduce our *observational* data (with errors) to compare with the theory and see just how all over the place it is. As a reminder, this observational data comes from the [Planck mission](http://sci.esa.int/planck/), which has fully mapped the CMB several times.

The cell below loads the observational data with accompanying error bars from a FITS file, which is the standard for observational astronomy. The FITS data cube is a way to contain all the information in one file, which is great if you're trying to download the data and don't want to miss anything, but can present a challenge since the file structure can get complicated. That's why the cell below has a lot more code in it. I have commented the code for clarification.

```python
from astropy.io import fits       # need new module for this part

hdulist = fits.open('./COM_PowerSpect_CMB_R2.02.fits')

binned_data = hdulist[7].data      # There's a lot of information in the data cube. We only want one 2d slice of it
hdulist.close()                    # Need this for memory allocation. Not the biggest deal for our tiny dataset, but a good habit. 

L_val = []                         # Empty arrays for the L min/max (x error bars) and the temperature (jyst called "err")
L_min = []
L_max = []
D_L = []
err = []


# For loop that breaks down the 2d slide into 1d strands. After this, we've turned a 3d data cube into a series of 1d arrays that we can work with
for i in range(len(binned_data)):           
    L_val.append(binned_data[i][0])
    L_min.append(binned_data[i][1])
    L_max.append(binned_data[i][2])
    D_L.append(binned_data[i][3])
    err.append(binned_data[i][4])

x_err_hi = []
x_err_lo = []

# The data only supplied absolute min/max errors, but matplotlib would prefer to have relative errors (relative to the center datapoint).
# We remedy this with simple subtraction, below.
for j in range(len(L_val)):
    x_err_hi.append(L_max[j]-L_val[j])
    x_err_lo.append(L_val[j]-L_min[j])

# Plot on its own just to make sure it works as intended
plt.errorbar(L_val, D_L, xerr=[x_err_lo, x_err_hi], yerr=err, linestyle='None', color='red', label='$Planck$')
```

### Plotting Theory and Observation
Now we plot the simulated data from before alongside the observational data. It should be immediately clear that all but one of these models does not fit the data, since the error bars are rather small.

```python
plt.errorbar(L_val, D_L, xerr=[x_err_lo, x_err_hi], yerr=err, linestyle='None', color='red', label='$Planck \ Obs.$')
plt.plot(axis, data_h0_0_40, label='$H_0$ = 45.0', color='gray', alpha=0.5)
plt.plot(axis, data_h0_0_50, label='$H_0$ = 50.0', color='gray', alpha=0.5)
plt.plot(axis, data_nominal, label='$H_0$ = 67.3 *', color='gray', alpha=0.5)
plt.plot(axis, data_h0_0_80, label='$H_0$ = 80.0', color='gray', alpha=0.5)
plt.plot(axis, data_h0_0_90, label='$H_0$ = 90.0', color='gray', alpha=0.5)

plt.legend(loc='upper right')
plt.xlabel('$\ell$')
plt.ylabel(r'$\frac {\ell (\ell+1)} {2 \pi}$ $C_\ell$  [$\mu$K$^2$]')
plt.title('Simulated and Observational Power Spectra')

#plt.savefig('./Wide-Varied-h_0-TT.pdf', bbox_inches='tight')
#plt.savefig('./Wide-Varied-h_0-TT.png', bbox_inches='tight', dpi=300)
```
