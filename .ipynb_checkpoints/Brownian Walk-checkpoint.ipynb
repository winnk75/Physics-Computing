## Background 
In 1827, Robert Brown observed random motion of particles within undisturbed water. He suspected that there was some physical law that governed this motion, but was ultimately unable to determine the cause. In 1905, Einstein provided an explanation; random impacts from water molecules (thousands and thousands per second) enact a truly stochastic force upon the large particle. __[This gif](https://commons.wikimedia.org/wiki/File:Brownian_motion_large.gif)__ shows a large particle (yellow) walking in a Brownian regime. It is impacted by many small particles (black) every second, which gives the particle some nonzero momentum.

To fully motivate this discussion, let us consider the fluctuation dissipation theorem. Consider a fluid with particles interacting with one another. As the particles collide, they dissipate energy as heat. And if this process takes place, the opposite must also take place; particles may also absorb heat as kinetic energy. This is Brownian Motion.

We will ultimately show that if the particles all begin at the origin, the distribution will spread over time as a gaussian with a standard deviation that is proportional to the square root of time (that is, $\sigma \propto t^{1/2}$).

## Derivation Sketch

### Motivations
Brownian motion comes from the predictable behavior of an ensemble of particles. We cannot predict the motion of any specific collision, but we *can* predict the average of many collisions. If the Brownian particle is getting bumped in all directions equally, we expect the average velocity over some time to be zero (it's going left half the time and right half the time), but the average *speed* to be nonzero (it's still moving left and right, not standing still). That is...

$\big<v(t)\big> \ = \ 0  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \      \big< \ \sqrt{v(t)^2} \ \big> \ \neq \ 0$

This is key. A net zero force on the particle (over long timescales) still imparts a force on the particle, giving it some momentum.

### Langevin Equation

Brownian Motion is best derived using the Langevin Equation, which models the diffusion using  Newtonian laws and a gaussian randomized correlation function.

$\frac{d V(t)}{dt} \ = \ - \frac{\gamma}{m} V(t) \ + \ \frac{1}{m} \ \epsilon (t)$

In this equation, $\gamma \ = \ 6 \pi \eta r$, and is the damping coefficient in accordance with Stokes Law ($\eta$ is the viscosity of the fluid, and $r$ is the radius of the diffusing particle. Stokes Law will not be derived here). The first two terms are straightforward: on the left side there is some acceleration, change in velocity with respect to time, and the first term on the right side is some damping term that scales with velocity, analagous to a damped harmonic oscillator. Essentially, the faster a particle is going, the more drag force there is slowing it down. The final term, $\epsilon (t)$, is an expression for the random impacts that change the particle. Because it’s random, there’s not really a functional form. Suffice to say that it's random in time, and that the amplitude of force provided is drawn normal distribution, with some known mean and standard deviation. This is much easier to express in code than it is using equations.

Skipping a whole lot of steps, we arrive at the Einstein Equation for Brownian Motion, below.

$D \ = \ \frac{k_B T}{\gamma} \ = \ \frac{k_B T}{6 \pi \eta a}$

This equation must also satisfy the diffusion equation, which is a differential equation.

$\frac{\partial \rho}{\partial t} \ = \ D \frac{\partial^2 \rho}{\partial x^2}$

We can "guess" that the solution is Gaussian, and plug in a Gaussian into this equation and see that it satisfies the conditions of the differential equation. It may be shown that the equation below is a solution to the differential equation, where $D$ is the diffusion constant provided above.

$\rho (x,t) \ = \ \frac{N}{\sqrt{4 \pi D t}} e^{- \frac{x^2}{4Dt}}$

This is a Gaussian function that starts at zero and spreads out in time. I'll introdce an analytical plot later, fow now let's get into the code!


```python
# Start with the basics. We need math and plots, so import them below.
# The last line allows plots to appear below the cells that generate them in Jupyter.

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

%matplotlib inline
```

```python
# Let's build a 2d random walk simulation from scratch.
# This is TRULY RANDOM WALK, not Brownian walk. That part is a little more complicated and we'll do it later.

n_steps = 20000     # Some number of iterations to perform
x_coord = 0.0       # starting location
y_coord = 0.0       # starting location
x_steps_array = [0.0]   # Array for our x coordinates. We start at zero and append every new location to see a path
y_steps_array = [0.0]   # Same thing for y coordinates.

i = 0
while i < n_steps:
    phi = 2*np.pi*np.random.rand()    # Some random phase between 0 and 2 pi.
    steplength = np.random.rand()     # Fully random number for now, may change easily
    
    x_coord = x_coord + steplength*np.cos(phi)    # Updates the location that we're at in x,y
    y_coord = y_coord + steplength*np.sin(phi)
    
    x_steps_array.append(x_coord)
    y_steps_array.append(y_coord)
    
    i=i+1

# Troubleshooting
#print(x_steps_array)
#print(y_steps_array)

# Trying to change the aspect ratio, but nothing works
#plt.axis('equal')
#plt.axes().set_aspect('equal', 'datalim')

plt.plot(x_steps_array,y_steps_array,linewidth=1, label="Random Walk")
plt.plot(x_steps_array[-1],y_steps_array[-1], linewidth=3, label="End Point", ls='none', marker='o')  # Paints the end value in orange.
plt.plot(x_steps_array[0],y_steps_array[0], linewidth=3, label="Start Point", ls='none', marker='o')

# Figures out the farthest divergence from zero in any direction, uses it to plot...
axes_limit = np.max( [np.max(x_steps_array), abs(np.min(x_steps_array)), np.max(y_steps_array), abs(np.min(y_steps_array))] )

plt.legend(loc='best')
plt.xlim( (-1.3*axes_limit, 1.3*axes_limit) )
plt.ylim( (-1.1*axes_limit, 1.1*axes_limit) )
```

## Truly Random Walk
The code above creates a truly random walk, *not* Brownian Motion. But it shows something very interesting. Even for a truly random process without any funny business, you *more often than not* end up pretty far from your initial starting point at zero.

Feel free to run the cell above a few times and see that it's very rare for this process to stay at zero, even if you're just as likely to go up or down, left or right, on any given step.

## Building a Brownian Motion Code
Now let's build something that uses Brownian Motion to step instead of random motion. We'll be going down to one dimension for this in order to save on computation time.

We ideally want to run our program a few hundred to a few thousand times in order to test the divergence of our simulation versus the theoretical divergence of the Gaussian we found in the introduction.

The cell below allows the user to set the number of runs, or walkers, (currently set to 1000) and the number of steps to take (currently set to 200). Each walker iterates through a number of gaussian jumps according to the number of steps.

With the default values, this should only take a few seconds, and will serve to show the diffusion effect we're interested in. Begin by running the cell below. For simplicity, this will only run the simulation and provide limited feedback. We plot the data in the following cell.

```python
# Let's try to make something vaguely brownian motion related

n_steps = 200     # Some number of iterations
n_runs = 1000
x_coord = np.zeros(n_runs)      # even though all runs start at x=0, the array makes more sense to dump arrays of random numners
x_steps_array = np.zeros( (n_runs,n_steps) )    # creates an array in advance for complexity reasons. the append() function is very slow for big computations.

sigma = 1.0         # Standard deviation for np.random.normal() function
mu = 0.0            # Mean for np.random.normal() function

print(len(x_steps_array[0]))

axis = np.arange(0,n_steps)

i = 0

while i < n_steps:
    if i == n_steps/2:
        print("halfway there!") # just a bit of feedback because this can take a long time
    j = 0
    while j < n_runs:
        steplength = np.random.normal(mu,sigma,n_runs)     # Fully random number for now, may change easily
    
        x_coord[j] = x_coord[j] + steplength[j]    # Updates the location that we're at
        x_steps_array[j][i] = x_coord[j]
        j = j+1  
    i=i+1
    
print("Simulation complete! Run the cell below to plot your results!")
```

```python
# Just the plotting function
print('Number of Steps: '+str(n_steps))
print('Number of Runs: '+str(n_runs))
print('Normal Distribution Median: '+str(mu))
print('Normal Distribution StDev: '+str(sigma))



k = 0
while k < n_runs:
    #plt.plot(axis,x_steps_array[k],linewidth=1)
    plt.plot(axis,x_steps_array[k],linewidth=1,color='gray',alpha=0.05)
    k=k+1
    
plt.title('Brownian Motion Walk, '+str(n_runs)+' runs, '+str(n_steps)+' steps, $\sigma$ = '+str(sigma))
plt.xlabel("Number of Jumps, or time")
plt.savefig('brownian-walk-gradient.png',bbox_inches='tight',dpi=600) # do a png, not a pdf
```

## Analysis of the Brownian Walk

The plot shows how every walker starts at zero and begins to drift. The average value of all the walkers is still zero (or close to it), but they've spread out over their various jumps. We can take vertical "slices" of the distribution at various times and plot these results as a histogram.

```python
# Analysis Section
halfway_sample = []
quarter_sample = []
threqtr_sample = []
final_sample= []
original_sample = []
tenth_sample = []

i = 0
while i < n_runs:
    tenth_sample.append(x_steps_array[i][int(round(n_steps/10,0))])
    halfway_sample.append(x_steps_array[i][int(round(n_steps/2,0))])
    quarter_sample.append(x_steps_array[i][int(round(n_steps/4,0))])   # takes the quarter-way-through distribution, with some sanitization steps that round to an intiger, then set the string back to int
    threqtr_sample.append(x_steps_array[i][int(round(0.75*n_steps,0))])
    final_sample.append(x_steps_array[i][-1])
    original_sample.append(x_steps_array[i][0])      # one step after initial conditions (first loop moves once) so not all zero
    i = i + 1

plt.hist(final_sample, label='1.00 t',bins=30)
#plt.hist(threqtr_sample, label='0.75 t',bins=10)
plt.hist(halfway_sample, label='0.50 t',bins=30)
plt.hist(quarter_sample, label='0.25 t',bins=30)
plt.hist(tenth_sample, label='0.10 t',bins=15)


plt.legend(loc='upper left')
plt.title('Histogram of walk distances at various timescales')
#plt.savefig('brownian-hist.png',bbox_inches='tight',dpi=300) # literally don't bother with pdf, it's 50+ mb


```

## Comparing the Simulation with Theory

This plot clearly illustrates how the distribution spreads out over time. We may also plot the probability distribution that we derived in the first cell of this notebook, and compare the theoretical distribution given by Einstein with the simulation distribution that we just plotted.

Run the cell below to plot the theoretical distrubution!

```python
tt = 190.
#tt = 42.
std_array = [(1.00*tt)**0.5, (0.50*tt)**0.5, (0.25*tt)**0.5, (0.10*tt)**0.5]

g0 = signal.gaussian(100, std=std_array[0])
g1 = signal.gaussian(100, std=std_array[1])
g2 = signal.gaussian(100, std=std_array[2])
g3 = signal.gaussian(100, std=std_array[3])


# Normalization constant = sqrt[ 2 * pi / sigma]
#norm0 = np.sqrt(2.*np.pi*1.) # normalization of a gaussian with sigma=1
norm0 = np.sqrt(2.*np.pi/std_array[0])
norm1 = np.sqrt(2.*np.pi/std_array[1])
norm2 = np.sqrt(2.*np.pi/std_array[2])
norm3 = np.sqrt(2.*np.pi/std_array[3])

axis = np.arange(-50,50)

plt.fill(axis,norm0*g0, alpha=1, label='1.00 t')
plt.fill(axis,norm1*g1, alpha=1, label='0.50 t')
plt.fill(axis,norm2*g2, alpha=1, label='0.25 t')
plt.fill(axis,norm3*g3, alpha=1, label='0.10 t')

plt.legend(loc='upper left')
plt.title('Theoretical distribution of walk distances at various timescales')

#plt.savefig('brownian-dist-10k.png', bbox_inches='tight', dpi=300)
```

## Looking Ahead: Geometric Brownian Motion and Financial Applications

My next project will take a long hard look at Geometric Brownian Motion (GBM), which is like basic Brownian Motion, but with some ever present "drift", a long term trend away from zero. This is a basic way to model stock markets: high volatility but a slight trend upwards given long enough time.

This [image](https://i.investopedia.com/inv/articles/site/CT-MonteCarlo3ra.gif) (from [this article](https://www.investopedia.com/articles/07/montecarlo.asp)) explains the concept rather well in my opinion.

The cell below uses 5 walkers (stocks) that fluctuate in a GBM regime for the number of steps, which is 2000. This can be 2000 minutes, hours, days, fortnights, or anything else. The point is that we have five walkers that behave with a strong volatility component, with a weaker drift component. I've also included an "ETF" which is just an unwighted mean of the five walkers. Note that the ETF curve follows the same drift model, but with a reduced standard deviation. This is why people diversify into multiple stocks.

I will be exploring financial applications in future notebooks, but for now I leave you with a proof of concept model, below.

```python
# Economics models proof of concept. Low n_runs here...
np.random.seed(38)       # Using a random seed here to illustrate the reduced variance of the "ETF". Obviously, this is cherry picked to show my exmaple.

n_steps = 2000     # Let's say days or something. Some time interval.
n_runs = 5         # Not too many
x_coord = np.zeros(n_runs)      # even though all runs start at x=0, the array makes more sense to dump arrays of random numners
x_steps_array = np.zeros( (n_runs,n_steps) )    # creates an array in advance for complexity
etf = np.zeros(n_steps)
etf_dummy_array = []

sigma = 1.0          # Standard deviation for np.random.normal() function
mu = 0.02            # Mean for np.random.normal() function


print(len(x_steps_array[0]))

axis = np.arange(0,n_steps)

i = 0        # i is the steps index, j is the runs index

while i < n_steps:
    if i == n_steps/2:
        print("halfway there!")
    j = 0
    while j < n_runs:
        steplength = np.random.normal(mu,sigma,n_runs)     # Fully random number for now, may change easily
    
        x_coord[j] = x_coord[j] + steplength[j]            # Updates the location that we're at
        x_steps_array[j][i] = x_coord[j]
       
        etf_dummy_array.append(x_steps_array[j][i])        # For a time, i, makes an array of all values, j.
        j = j+1
            
    etf[i] = np.average(etf_dummy_array)    # Averages the dummy array into our "ETF" with even weighting
    etf_dummy_array = []                    # Clears the dummy array for the next time, i.
    
    i=i+1
    

k = 0
while k < n_runs:
    plt.plot(axis,x_steps_array[k],linewidth=0.5,alpha=0.4,color='gray',label='Security '+str(k+1))
    k=k+1

plt.plot(axis,etf,color='Navy',label='Winn\'s ETF')
plt.legend(loc='upper left')

```

```python
# An alternate coding of GBM, this time from wikipedia
# Code from wikipedia https://en.wikipedia.org/wiki/Geometric_Brownian_motion


# Winn's interpretation: The drift parameter maintains a steady march in one way or the other,
# ...but the variance, sigma, can alter this. For the high sigma values, sigma completely dominates the function,
# ...leaving a net zero. While the low sigma value experience a higher drift value
#python code for the plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

mu=1
n=50
dt=0.1
x0=100
x=pd.DataFrame()
np.random.seed(1)

for sigma in np.arange(0.8,2,0.2):
    step=np.exp((mu-sigma**2/2)*dt)*np.exp(sigma*np.random.normal(0,dt,(1,n)))
    temp=pd.DataFrame(x0*step.cumprod())
    x=pd.concat([x,temp],axis=1)

x.columns=np.arange(0.8,2,0.2)
plt.plot(x)
plt.legend(x.columns)
plt.xlabel('t')
plt.ylabel('X')
plt.title('Realizations of Geometric Brownian Motion with different variances\n mu=1')
plt.show()
```
